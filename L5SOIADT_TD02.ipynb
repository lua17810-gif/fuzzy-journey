```python
import glob
print(glob.glob("DATA/*"))
```

```python
import matplotlib.pyplot as pyplot      

for chemin in glob.glob("DATA/*"):
    print(chemin)
    toto = open(chemin,encoding="utf-8")
    titi=toto.read()
    toto.close()
    tata = titi.split()
    tutu = {}
    for mot in tata:
        if mot not in tutu:
            tutu[mot]=1
        else:
            tutu[mot]+=1
    liste_effectifs = sorted(list(tutu.values()), reverse=True)
    pyplot.plot(liste_effectifs)
pyplot.yscale('log')
pyplot.xscale('log')
pyplot.show()
```
