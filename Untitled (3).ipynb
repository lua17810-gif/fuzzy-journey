```python
from typing import List

def nettoyer_texte(texte: str) -> List[str]:
    texte_min=texte.lower()
    texte_min = texte_min.replace('\n',' ')
    caracteres_speciaux = '!?,;:."()[]{}<>*#@$%^&~'
    for char in caracteres_speciaux:
        texte_min = texte_min.replace(char, '')
    
    mots = texte_min.split()
    return mots
```
