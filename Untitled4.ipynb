```python
import glob
import json

print(glob.glob("C:\\Users\\32534\\Music\\corpus_multi\\*"))
dic_langue = {}

for chemin in glob.glob("C:\\Users\\32534\\Music\\corpus_multi\\*\\appr\\*"):
    dossiers = chemin.split("\\")
    langue = dossiers[5]
    
    
    if langue not in dic_langue:
        dic_langue[langue] = {}
    
    with open(chemin, "r", encoding="utf-8") as f:
        chaine = f.read()
        mots = chaine.split()
        for m in mots:
            if m not in dic_langue[langue]:
                dic_langue[langue][m] = 1
            else:
                dic_langue[langue][m] += 1

print(dic_langue)

dic_modeles = {}
for langue, dic_effectifs in dic_langue.items():
    paires = [[effectif, mot] for mot, effectif in dic_effectifs.items()]
    liste_tri = sorted(paires)[-10:]
    dic_modeles[langue] = {mot:effectif for effectif, mot in liste_tri}

with open("modeles.json", "w", encoding="utf-8") as w:
    w.write(json.dumps(dic_modeles, indent=2))
    
```

```python
langues_mots = []
for langue, mots in dic_modeles.items():
    langues_mots.append((langue, set(mots.keys())))
    print(f"\n{langue}: {set(mots.keys())}")
```
