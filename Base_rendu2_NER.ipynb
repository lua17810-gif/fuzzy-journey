```python
def ouvrir_fichier(path):
    with open(path, encoding="utf-8") as f:
        chaine = f.read()
    return chaine
import spacy
nlp = spacy.load("fr_core_news_sm")

for path in glob.glob("corpus-REN_2023-2024/*/*.txt"):
    print(path)
    liste_entites= []
    texte = ouvrir_fichier(path)
    #doc = nlp(texte[:1000])
    doc = nlp(texte)
    for ent in doc.ents:
        #print(ent.text, ent.label_)
        liste_entites.append(ent.text)
    break
```
