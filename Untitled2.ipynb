```python
for chemin in glob.glob("corpus_multi\\*\\appr\\*"):
    dossiers= chemin.split('\\')
    langue = dossiers[1]
    print(langue)
    dic_langue={}
    for langue in dossiers:
        if langue not in dic_langue:
            dic_langue[langue]={}
    f=open(chemin,"r",encoding="utf-8")
    chaine=f.read()  
    mots=chaine.split() 
    for m in mots:
        if m not in dic_langue[langue]:
            dic_langue[langue][m]=1
        else:
            dic_langue[langue][m]+=1
print(dic_langue)
```
