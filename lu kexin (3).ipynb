```python
import glob
import json
import os


def lire_fichier(chemin):
    """Lit un fichier texte et renvoie son contenu."""
    with open(chemin, "r", encoding="utf-8") as f:
        return f.read()


def charger_modele(chemin_modele):
    """Charge le modèle JSON enregistré."""
    with open(chemin_modele, "r", encoding="utf-8") as f:
        return json.load(f)


def predire_langue(texte, modeles, top_n=10):
    """Prédit la langue d'un texte à partir des modèles de langue."""
    mots = texte.split()
    dic_freq = {}
    for m in mots:
        dic_freq[m] = dic_freq.get(m, 0) + 1

    paires = [[eff, mot] for mot, eff in dic_freq.items()]
    paires_triees = sorted(paires)[-top_n:]
    plus_frequents = set([mot for eff, mot in paires_triees])

    liste_predictions = []
    for langue_ref, model in modeles.items():
        mots_communs = set(model).intersection(plus_frequents)
        nb_mots_communs = len(mots_communs)
        liste_predictions.append([nb_mots_communs, langue_ref])

    meilleure_prediction = sorted(liste_predictions, reverse=True)[0][1]
    return meilleure_prediction


def evaluer(chemin_corpus, chemin_modele, top_n=10):
    """Évalue le système : exactitude, rappel, précision et F1 par langue."""
    modeles = charger_modele(chemin_modele)
    fichiers_test = glob.glob(os.path.join(chemin_corpus, "*", "test", "*"))
    separateur = "\\" if os.name == "nt" else "/"

    nb_bonnes = 0
    total = len(fichiers_test)
    dic_resultats = {}

    if total == 0:
        print("Aucun fichier de test trouvé. Vérifiez le chemin du corpus.")
        return

    for chemin in fichiers_test:
        dossiers = chemin.split(separateur)
        langue_reelle = dossiers[1]
        texte = lire_fichier(chemin)
        langue_predite = predire_langue(texte, modeles, top_n)

        dic_resultats.setdefault(langue_reelle, {"VP": 0, "FP": 0, "FN": 0})
        dic_resultats.setdefault(langue_predite, {"VP": 0, "FP": 0, "FN": 0})

        if langue_predite == langue_reelle:
            nb_bonnes += 1
            dic_resultats[langue_reelle]["VP"] += 1
        else:
            dic_resultats[langue_reelle]["FN"] += 1
            dic_resultats[langue_predite]["FP"] += 1

    exactitude = nb_bonnes / total
    print(f"\n Évaluation du modèle ({total} fichiers testés)")
    print(f" Exactitude globale : {exactitude:.2%}\n")

    print(f"{'Langue':<10} {'VP':>4} {'FP':>4} {'FN':>4} | {'Rappel':>8} {'Précision':>10} {'F1':>8}")
    print("-" * 60)
    for langue, stats in dic_resultats.items():
        vp, fp, fn = stats["VP"], stats["FP"], stats["FN"]
        rappel = vp / (vp + fn) if (vp + fn) > 0 else 0.0
        precision = vp / (vp + fp) if (vp + fp) > 0 else 0.0
        f1 = (2 * rappel * precision / (rappel + precision)) if (rappel + precision) > 0 else 0.0

        print(f"{langue:<10} {vp:>4} {fp:>4} {fn:>4} | {rappel:>8.2%} {precision:>10.2%} {f1:>8.2%}")

    return dic_resultats, exactitude


if __name__ == "__main__":
    evaluer( chemin_corpus=r"C://Users//32534//Music//corpus_multi",chemin_modele="models.json",top_n=10 )
```

```python
import glob
import json
import os
from collections import Counter


def lire_fichier(chemin):
    """Lit un fichier texte et renvoie son contenu."""
    with open(chemin, "r", encoding="utf-8") as f:
        return f.read()


def extraire_ngrams(texte, n=3):
    """Découpe un texte en n-grammes de caractères."""
    texte = texte.replace("\n", " ").replace("\r", " ")
    return [texte[i:i+n] for i in range(len(texte)-n+1)] if len(texte) >= n else []


def creer_modeles_caracteres(corpus_path="corpus_multi", n=3, top_k=10):
    """Crée les modèles de langue basés sur des n-grammes de caractères."""
    dic_langues = {}

    fichiers_appr = glob.glob(os.path.join(corpus_path, "*", "appr", "*"))
    print(f"Nombre de fichiers d'apprentissage : {len(fichiers_appr)}")

    for chemin in fichiers_appr:
        dossiers = chemin.split("\\") if "\\" in chemin else chemin.split("/")
        langue = dossiers[1]
        texte = lire_fichier(chemin)
        ngrams = extraire_ngrams(texte, n)

        if langue not in dic_langues:
            dic_langues[langue] = Counter()

        dic_langues[langue].update(ngrams)

    dic_modeles = {}
    for langue, compteur in dic_langues.items():
        top = [ng for ng, freq in compteur.most_common(top_k)]
        dic_modeles[langue] = top

    with open(f"models_char_{n}.json", "w", encoding="utf-8") as f:
        json.dump(dic_modeles, f, indent=2, ensure_ascii=False)

    print(f" Modèles de langue (caractères, n={n}) créés → models_char_{n}.json")
    return dic_modeles


def charger_modele(chemin_modele):
    with open(chemin_modele, "r", encoding="utf-8") as f:
        return json.load(f)


def predire_langue_caracteres(texte, modeles, n=3, top_k=10):
    """Prédit la langue d’un texte via les n-grammes de caractères."""
    ngrams = extraire_ngrams(texte, n)
    compteur = Counter(ngrams)
    plus_frequents = set([ng for ng, _ in compteur.most_common(top_k)])

    predictions = []
    for langue_ref, model in modeles.items():
        communs = set(model).intersection(plus_frequents)
        predictions.append([len(communs), langue_ref])

    meilleure_langue = sorted(predictions, reverse=True)[0][1]
    return meilleure_langue


def evaluer_caracteres(corpus_path="corpus_multi", chemin_modele="models_char_3.json", n=3, top_k=10):
    """Évalue les modèles à base de n-grammes de caractères."""
    modeles = charger_modele(chemin_modele)
    fichiers_test = glob.glob(os.path.join(corpus_path, "*", "test", "*"))
    separateur = "\\" if os.name == "nt" else "/"

    total = len(fichiers_test)
    nb_bonnes = 0
    dic_resultats = {}

    for chemin in fichiers_test:
        dossiers = chemin.split(separateur)
        langue_reelle = dossiers[1]
        texte = lire_fichier(chemin)
        langue_predite = predire_langue_caracteres(texte, modeles, n, top_k)

        dic_resultats.setdefault(langue_reelle, {"VP": 0, "FP": 0, "FN": 0})
        dic_resultats.setdefault(langue_predite, {"VP": 0, "FP": 0, "FN": 0})

        if langue_predite == langue_reelle:
            nb_bonnes += 1
            dic_resultats[langue_reelle]["VP"] += 1
        else:
            dic_resultats[langue_reelle]["FN"] += 1
            dic_resultats[langue_predite]["FP"] += 1

    exactitude = nb_bonnes / total if total > 0 else 0

    print(f"\n Évaluation du modèle à base de caractères (n={n})")
    print(f"Exactitude globale : {exactitude:.2%}\n")

    print(f"{'Langue':<10} {'VP':>4} {'FP':>4} {'FN':>4} | {'Rappel':>8} {'Précision':>10} {'F1':>8}")
    print("-" * 60)
    for langue, stats in dic_resultats.items():
        vp, fp, fn = stats["VP"], stats["FP"], stats["FN"]
        rappel = vp / (vp + fn) if (vp + fn) > 0 else 0
        precision = vp / (vp + fp) if (vp + fp) > 0 else 0
        f1 = (2 * rappel * precision / (rappel + precision)) if (rappel + precision) > 0 else 0
        print(f"{langue:<10} {vp:>4} {fp:>4} {fn:>4} | {rappel:>8.2%} {precision:>10.2%} {f1:>8.2%}")

    return dic_resultats, exactitude


if __name__ == "__main__":
    
    creer_modeles_caracteres(corpus_path=r"C://Users//32534//Music//corpus_multi", n=3,top_k=10)

    evaluer_caracteres(corpus_path=r"C://Users//32534//Music//corpus_multi",chemin_modele="models_char_3.json",n=3,top_k=10)
```
